```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_test.csv")
```

```python
display(df)
```

```python
import matplotlib.pyplot as plt

plt.stem(df['households'][:20], df['population'][:20])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.stackplot(df.index[:20], df['housing_median_age'][:20], df['total_rooms'][:20], df['total_bedrooms'][:20])
plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Stack Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(df['latitude'][:20], df['longitude'][:20])
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Line Plot of Latitude vs Longitude")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.bar(df.index[:20], df['median_income'][:20])
plt.xlabel("Index")
plt.ylabel("Median Income")
plt.title("Bar Plot of Median Income")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = df.index[:5]
sizes = df['housing_median_age'][:5]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title("Pie Chart of Housing Median Age (First 5 entries)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.rugplot(df['median_income'][:100])
plt.title("Rug Plot of Median Income (First 100 entries)")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['latitude'][:100], df['longitude'][:100])
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Scatter Plot of Latitude vs Longitude (First 100 entries)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df['age_group'] = pd.cut(df['housing_median_age'], bins=5)

sns.boxplot(x=df['age_group'][:100], y=df['median_income'][:100])
plt.title("Box Plot of Median Income by Housing Age Group (First 100 entries)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.swarmplot(x=df['age_group'][:100], y=df['median_income'][:100])
plt.title("Swarm Plot of Median Income by Housing Age Group (First 100 entries)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

df_agg = df.groupby('age_group')['median_income'].median().reset_index()

plt.bar(df_agg['age_group'].astype(str), df_agg['median_income'])
plt.xlabel("Age Group")
plt.ylabel("Median Income")
plt.title("Bar Plot of Median Income by Housing Age Group")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

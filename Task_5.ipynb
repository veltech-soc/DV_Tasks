```python
import plotly.express as px
import pandas as pd

data = {
    'category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'subcategory': ['A1', 'A2', 'B1', 'B2', 'C1', 'C2'],
    'value': [10, 15, 20, 12, 8, 18]
}
df = pd.DataFrame(data)

fig_treemap = px.treemap(df, path=['category', 'subcategory'], values='value',
                         title='Treemap with Rectangular Layout ')
fig_treemap.show()

fig_sunburst = px.sunburst(df, path=['category', 'subcategory'], values='value',
                           title='Sunburst Chart ')
fig_sunburst.show()
```

```python
import pandas as pd
import plotly.express as px

education_df = pd.read_csv('/root/.cache/kagglehub/datasets/adilshamim8/educational-backgrounds-of-successful-people/versions/2/successful_educations.csv')

print("First 5 rows of the dataset:")
display(education_df.head())
print("\nColumns in the dataset:")
print(education_df.columns)

education_counts = education_df.groupby(['Profession', 'Degree']).size().reset_index(name='counts')

fig_treemap_edu = px.treemap(education_counts, path=['Profession', 'Degree'], values='counts',
                             title='Treemap of Educational Backgrounds by Profession and Degree')
fig_treemap_edu.show()

fig_sunburst_edu = px.sunburst(education_counts, path=['Profession', 'Degree'], values='counts',
                              title='Sunburst Chart of Educational Backgrounds by Profession and Degree')
fig_sunburst_edu.show()
```

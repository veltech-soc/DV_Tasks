```python
print("Column names and data types:")
display(df.info())

print("\nSummary statistics of numerical columns:")
display(df.describe())

print("\nSummary statistics of categorical columns:")
display(df.describe(include='object'))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

selected_cols = ['x3', 'x4', 'x5', 'x6', 'y']
sns.pairplot(df[selected_cols].dropna())
plt.suptitle('Pair Plot of Selected Variables', y=1.02)
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.dropna(subset=['x4', 'y', 'x2']), x='x4', y='y', hue='x2', alpha=0.6)
plt.title('Scatter Plot of x4 vs y colored by x2')
plt.xlabel('x4')
plt.ylabel('y')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df.dropna(subset=['x2', 'y']), x='x2', y='y')
plt.title('Distribution of y by x2')
plt.xlabel('x2')
plt.ylabel('y')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(data=df.dropna(subset=['x2', 'y']), x='x2', y='y', palette='pastel')
plt.title('Distribution of y by x2')
plt.xlabel('x2')
plt.ylabel('y')
plt.xticks(rotation=90)
plt.show()
```

```python
from pandas.plotting import parallel_coordinates
from pandas.plotting import andrews_curves


df_numerical = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(12, 8))
parallel_coordinates(df_numerical.dropna(), 'id', color=sns.color_palette())
plt.title('Parallel Coordinates Plot')
plt.show()

plt.figure(figsize=(12, 8))
andrews_curves(df_numerical.dropna(), 'id', color=sns.color_palette())
plt.title("Andrew's Curves")
plt.show()
```

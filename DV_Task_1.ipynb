```python
import seaborn as sns
df = sns.load_dataset('mpg')
display(df.head())
```

```python
display(df.info())
display(df.isnull().sum())
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
numerical_cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year']
numerical_df = df[numerical_cols]
display(numerical_df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.boxplot(numerical_df['mpg'])
plt.title('Distribution of MPG')
plt.ylabel('MPG')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(y=df['mpg'])
plt.title('Distribution of MPG using Seaborn')
plt.ylabel('MPG')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='weight', y='mpg', data=df)
plt.title('Scatter Plot of Weight vs. MPG')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='horsepower', kde=True)
plt.title('Distribution of Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Frequency')
plt.show()
```
